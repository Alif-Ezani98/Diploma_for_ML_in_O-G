## Assignment 2: Write a code, where porosity and permeability thresholds will be given by user and you will return final dataframe.

```python
import numpy as np
import pandas as pd
```

```python
depth = np.arange(1500,5400,200)
porosity = np.array([0.15,0.13,0.18,0.24,0.23,0.29,0.33,0.35,0.43,0.39,0.37,0.45,0.36,0.43,0.33,0.42,0.22,0.24,0.22,0.28])
permeability = [20,34,25,58,79,98,150,190,230,198,224,216,275,213,237,389,98,86,62,54]

field_data = {"Depth" : depth, "Porosity" : porosity, "Permeability" : permeability}
df = pd.DataFrame(field_data)
```

```python
df
```

```python
poro_thresh = float(input("What is the porosity threshold? : "))
perm_thresh = float(input("What is the permeability threshold? : "))
print()

print(f"The zone(s) with porosity < {poro_thresh} and permeability < {perm_thresh} is/are as following:")

df[(df["Porosity"] < poro_thresh) & (df["Permeability"] < perm_thresh)]
```
